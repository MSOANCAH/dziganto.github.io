### Objective

This notebook shows how to accomplish several image processing tasks that are useful when plying deep learning.

### Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
import skimage.measure
```

### Function

```python
def img_info(ndarray, gray=0):
    '''displays image and shape; gray is a flag used by matplotlib'''
    if gray:
        plt.imshow(ndarray, cmap = plt.get_cmap('gray'));
    else:
        plt.imshow(ndarray);
    print('Image shape:', ndarray.shape)
```

### Set Image Path

Simply put, this is the path to the image or images you want to read in.

```python
img_path = '/Users/davidziganto/Desktop/lena.png'
```

### Get Image

```python
img = misc.imread(img_path)[...,:3]
```

### Show Original Image & Dimensions

```python
img_info(img)
```

### Convert to Grayscale

This is useful for Convolutional Neural Networks. Consider how much processing is eliminated by going from a 3-channel RGB image to a 1-channel Grayscale image.

```python
# Using Matlab's formula
grayscale = np.dot(img, [0.2989, 0.5870, 0.1140])
```

```python
img_info(grayscale, gray=1)
```

### Downsample Image

This too is useful for Convolutional Neural Networks. We can keep the most salient features while substantially decreasing computation.

```python
downsample = skimage.measure.block_reduce(grayscale, (2,2), np.max)
```

```python
img_info(downsample, gray=1)
```

### Standardize Data

This is a must any time you plan to leverage Gradient Descent. 

```python
stats = [np.min, np.max, np.mean, np.std]
for stat in stats:
    print( str(stat), stat(grayscale) )
```

```python
standardize = (downsample - downsample.mean()) / np.sqrt(downsample.var() + 1e-5)
```

```python
for stat in stats:
    print( str(stat), stat(standardize) )
```

```python
img_info(standardize, gray=1)
```
